```python
# BM = df["make"] == "BMW"
true_true_love = pd.DataFrame(columns=df.columns)
print(true_true_love)
i=0
for a in column:
    true_love = df[df["make"] == a]
#     print(true_love)
    
    true_true_love = pd.concat([true_true_love,true_love])
print(true_true_love)
true_true_love.to_csv('selected_matadata.csv')
```
